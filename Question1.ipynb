```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data=pd.read_csv(r'city_day.csv')
deaths=pd.read_csv(r'Book1.csv')
```

```python
names=[]
len(data)
```

```python
deaths
```

```python
data=pd.DataFrame(data)
```

```python
data['BTX']=data['Benzene']+data['Toluene']+data['Xylene']
```

```python
for i in range(len(data)):
    if data.iloc[i]['City'] not in names:
        
    
      
        names.append(data.iloc[i]['City'])
```

```python
names
```

```python
data.shape
```

```python
dataf = data.dropna(subset=['AQI'])
```

```python
AQI = []

for city in names:
    count = 0
    mass = 0
    for i in range(len(dataf)):
        if dataf.iloc[i]['City'] == city and pd.notna(dataf.iloc[i]['AQI']) and dataf.iloc[i]['AQI'] != 0:
            mass += dataf.iloc[i]['AQI']  # Ensure you're using 'dataf' consistently
            count += 1
    
    if count > 0:
        AQI.append(mass / count)  # Append the average AQI
    else:
        AQI.append(0)  # Append 0 (or another default value) if no valid AQI is found

            
            
         
    
```

```python
# now we will plot the AQI level and the no of deaths due to air pollution
len(deaths)
```

```python
deaths=pd.DataFrame(deaths)
```

```python
deathsdata=[]
for i in range(len(deaths)):
    deathsdata.append(deaths.iloc[i]['Deaths'])
    
```

```python
deathsdata
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.plot(deathsdata, AQI, marker='o', linestyle='-', color='b', linewidth=2, markersize=8, label='AQI vs Deaths')


plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)


plt.title('Relationship Between Number of Deaths and AQI', fontsize=16, fontweight='bold')
plt.xlabel('Number of Deaths', fontsize=14)
plt.ylabel('Air Quality Index (AQI)', fontsize=14)


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


plt.legend(fontsize=12, loc='upper left')


for i, (x, y) in enumerate(zip(deathsdata, AQI)):
    if i % 3 == 0:  # Annotate every 3rd point
        plt.annotate(f'({int(x)}, {int(y)})', 
                     (x, y), 
                     textcoords="offset points", 
                     xytext=(10, -10), 
                     fontsize=10, 
                     arrowprops=dict(arrowstyle='->', color='gray', lw=0.5))


plt.tight_layout()
plt.show()
```
